```python
import csv
books2 = [
    ['title', 'author' , 'year'],
    ['The Weirdstone of Brisingamen', 'Alan Garner', '1960'],
    ['Perdido Street Station', 'China Miéville', '2000'],
    ['Thud!', 'Terry Pratchett', '2005'],
    ['The Spellman Files', 'Lisa Lutz', '2007'],
    ['Small Gods', 'Terry Pratchett', '1992'],
    ]

with open('books2.csv', 'wt') as fout:
    csvout = csv.writer(fout)
    csvout.writerows(books2)
```

```python
import sqlite3
import csv
import sqlalchemy as sa

conn = sqlite3.connect('books.db')
curs = conn.cursor()

curs.execute('''CREATE TABLE IF NOT EXISTS books(
    title VARCHAR(30) PRIMARY KEY,
    author VARCHAR(20),
    year INT)''')
conn.commit()

curs.close()
conn.close()

conn = sqlite3.connect('books.db')
curs = conn.cursor()

with open("books2.csv", "rt") as books2:
    reader = csv.reader(books2)
    rows = list(reader)
    
    if rows:
        if rows[0][0].lower() in ['title', 'name', 'book']:
            rows = rows[1:]
        
        valid_rows = [row for row in rows if len(row) == 3]
        curs.executemany("INSERT INTO books VALUES(?,?,?)", valid_rows)
    else:
        print("No data found in CSV file")
        
conn.commit()

curs.close()
conn.close()
```

```python
import sqlite3
import sqlalchemy as sa
from sqlalchemy import text

conn = sa.create_engine('sqlite:///books.db')

with conn.connect() as conn:
    
    rows = conn.execute(text('SELECT title from books ORDER BY title ASC'))

    for row in rows:
        print(row)
        
conn.commit()
conn.close()
```
