```python
from datetime import date
import time

today = str(date.today())
with open ("today.txt", "w") as text_file:
    text_file.write(now)

with open ("today.txt", "rb") as text_file:
    today_string = text_file.read()

print(today_string)

fmt = "%Y %m %d"
time.strptime("today_string", fmt)
```
