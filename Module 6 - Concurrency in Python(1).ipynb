```python
from datetime import date
import time

today = str(date.today())
with open ("today.txt", "w") as text_file:
    text_file.write(today)

with open ("today.txt", "r") as text_file:
    today_string = text_file.read()

print(today_string)

fmt = "%Y-%m-%d"
time.strptime(today_string, fmt)
```

```python
'''
Use multiprocessing to create three separate processes. 
Make each one wait a random number of seconds between zero and one, 
print the current time, and then exit.
'''

import multiprocessing
import time
import random

def worker_process():
    time.sleep(random.random())
    print(datetime.now())

if __name__ == '__main__':
    processes = []
    for i in range(3):
        p = multiprocessing.Process(target=worker_process)
        processes.append(p)
        p.start()

    for p in processes:
        p.join()
```
